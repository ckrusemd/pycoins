# Best Opportunity

```python
import pandas as pd
from IPython.display import display, HTML
df = pd.read_csv('fit_data.csv')
pd.set_option('display.float_format', lambda x: '%.0f' % x)
```

```python
display(HTML(df[df.groupby('Symbol').Date.transform('max') == df['Date']].sort_values('Best_Fit',ascending=False).round(0).to_html()))
```
